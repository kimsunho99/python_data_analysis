```python
selloutData["REGIONID"]=selloutData["REGIONID"].astype(str)
selloutData["PRODUCT"]=selloutData["PRODUCT"].astype(str)
selloutData["YEARWEEK"]=selloutData["YEARWEEK"].astype(str)
selloutData["QTY"]=selloutData["QTY"].astype(float)
#형변환 정답은?
```

### 2. [불량 데이터 처리] kopo_channel_seasonality_new.csv 자료를 담은 selloutData 변수에서 QTY컬럼 음수(반품)인 경우 0, 양수인 경우 기존 QTY 값 유지하는 로직을 적용하여 QTY_NEW 컬럼을 추가하세요

```python
import pandas as pd
import numpy as np
```

```python
selloutData["QTY_NEW"] = np.where(selloutData["QTY"] < 0, 0, selloutData["QTY"])
```

```python
selloutData
```

```python
selloutData[selloutData.QTY < 0]["QTY"].count()
```

### 3. [데이터타입 통합] selloutData 자료에서 YEAR, WEEK 컬럼을 생성하고 WEEK 컬럼 값이 52 이하인 데이터만 refinedSelloutData 변수에 저장하세요

```python
selloutData["YEAR"] = selloutData["YEARWEEK"].str[0:4]
selloutData["WEEK"] = selloutData["YEARWEEK"].str[4:]
```

```python
selloutData.dtypes
```

```python
refinedSelloutData = selloutData[selloutData["WEEK"].astype(int) <= 52]
```

```python
refinedSelloutData.head()
```

### 4. refinedSelloutData 에서 지역, 상품, 연주차 컬럼순으로 오름차순으로 정렬하여 sortedData 변수에 담으세요

```python
sortedData = refinedSelloutData.sort_values(by=['REGIONID', 'PRODUCT', 'YEARWEEK'], ascending=True)

# 정렬된 데이터 출력
```

```python
sortedData
```

### 5. [지역, 상품, 연도 별 집계] sortedData 에서 지역, 상품, 연도 단위 판매량(QTY_NEW)의 평균 연산 후 groupData 변수에 담으세요. 이후 컬럼명을 QTY_MEAN로 변경하세요

```python
groupKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
groupData = sortedData.groupby(groupKey, as_index=False).QTY_NEW.mean()
```

```python
groupData = groupData.rename(columns = {'QTY_NEW':'QTY_MEAN'})
```

### 6. refinedSelloutData와 groupData를 [REGIONID, PRODUCT, YEAR] 키로 조인하여 mergedData 변수에 아래와 같이 담으세요

```python
joinKey = ["REGIONID","PRODUCT","YEAR"]
```

```python
mergedData = pd.merge(left = refinedSelloutData, right = groupData, on = joinKey, how ="left")
mergedData.head()
```

###7. mergedData를 활용하여 SEASONALITY = QTY_NEW / QTY_MEAN SEASONALITY 컬럼을 생성한다

```python
SEASONALITY = mergedData["SEASONALITY"] = mergedData.QTY_NEW/mergedData.QTY_MEAN
```

```python
mergedData
```

### 8.마지막으로 [지역, 상품, 주차] 별 SEASONALITY 평균값을 산출하여 finalResult 변수에 담으세요

```python
finalResult  = mergedData.groupby(by=["REGIONID","PRODUCT","WEEK"])["SEASONALITY"].agg(["mean"]).reset_index()
```

```python
finalResult
```

```python
finalResult.rename(columns = {'mean':'seasonality'})
```
